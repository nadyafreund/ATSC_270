### DOWNLOAD OBSERVATIONS FROM  FILE ###
RAwinsonde OBservations (RAOBs) are available from the Iowa State Environmental Mesonet (IEM) archive.  This data can be accessed using the siphon module.
```python
#grab data from 00Z on 26 December 2023
dt=datetime(2023,12,26,0)
get_raobs(dt)
```

### SUBSET TO 500MB OBS ###
We'll also do this within a function in the assignment_5_functions file.

```python
press=select_press(get_raobs(dt),500)
print(press)
```

### PLOT DATA MAP ###
Use the same map of CONUS as we did for the beginning of surface obs.<br>
No need to thin obs as there aren't a ton of RAOBs.

```python
#map projection using a Lambert Conformal Projection
#we'll center the map at 100W, 35N
proj=ccrs.LambertConformal(central_longitude=-100,central_latitude=35)

#create 10" x 10" figure with a dpi=100
fig=plt.figure(figsize=(10,10),dpi=100)

#add subplot with projection
ax=fig.add_subplot(1,1,1,projection=proj)

#add elements to make map more useable
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.STATES)
ax.add_feature(cfeature.BORDERS)

#zoom on the CONUS
#lon_min= -120
#lon_max= -75
#lat_min= 23
#lat_max= 50
ax.set_extent((-120,-75,23,50))

pressure=press['pressure'].values*units('millibar')
height=press['height'].values*units('meters')
temperature=press['temperature'].values*units('degC')
dewpoint=press['dewpoint'].values*units('degC')
direction=press['direction'].values*units('degree')
speed=press['speed'].values*units('m/s')
station=press['station']
time=press['time'].values
u_wind=press['u_wind'].values
v_wind=press['v_wind'].values
u,v=mpcalc.wind_components(press['speed'].values*units('m/s'),press['direction'].values*units('degree'))
latitude=press['latitude'].values
logitude=press['longitude'].values

plt.scatter(press.longitude,press.latitude,transform=ccrs.PlateCarree(),color='black')
#specify the axes to draw on, as well as the lon/lat of the stations (with transform)
stationplot=StationPlot(ax,press['longitude'],press['latitude'],transform=ccrs.PlateCarree(),fontsize=9)

#temperature in red
stationplot.plot_parameter('NW',press['temperature'].values,color='red')
#dewpoint DEPRESSION in dark green          #################################
stationplot.plot_parameter('SW',press['dewpoint'].values,color='darkgreen')
#height (need to format to be the first 3 digits) ###########################
stationplot.plot_parameter((2,1),press['height'].values)
#station ID in dark blue
stationplot.plot_text((2,0),press['station'].values,color='darkblue')
#wind in black
stationplot.plot_barb(u,v,color='black')

#save the map as upper_air_500.png
fig.tight_layout()
plt.savefig('upper_air_500.png')
```
