```python
#importing required libraries

import pandas as pd
import numpy as np
from datetime import datetime
from metpy.io import metar
import metpy.calc as mpcalc
from metpy.calc import wind_components
from metpy.units import units
from metpy.plots import Hodograph, SkewT
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from functions import get_raobs
```

```python
#open the file

#EF4 Tornado in Dayton, OH occured @ 22:31 EDT @ 05/27/2019
#EF4 Tornado in Dayton, OH occured @ 02:31 UTC @ 05/28/2019
#05/28/2019 @ 00:00 UTC
sounding_file=open('KILN_2019052800_2019052810.txt','r').readlines()
```

```python
#reading data into lists

#append works best here, even though appending is bad
#creating empty lists to hold sounding values

#3 or 4 character station identifier
station=[]
#RAOB launch cycle time in UTC
validUTC=[]
#level code for this observation
#4=mandatory level
#5=significant level
#6=wind level (PPBB)
#7=tropopause level
#8=maximum wind level
#9=surface level
levelcode=[]
#pressure in millibars
pressure_mb=[]
#height above sea level in meters
height_m=[]
#air temperature in celsius
tmpc=[]
#dew point temperature in celsius
dwpc=[]
#wind direction in degrees from North
drct=[]
#wind speed in knots
speed_kts=[]
#balloon bearing in degrees
bearing=[]
#ballon range from launch in statute miles
range_sm=[]

#loop through each line in the sounding, skipping the header (first line)
for line in sounding_file[1:]:
    #appending to each list the variable
    #station is the zero-th column (split by commas)
    station.append(line.split(",")[0])
    validUTC.append(line.split(",")[1])
    levelcode.append(line.split(",")[2])
    pressure_mb.append(line.split(",")[3])
    height_m.append(line.split(",")[4])
    tmpc.append(line.split(",")[5])
    dwpc.append(line.split(",")[6])
    drct.append(line.split(",")[7])
    speed_kts.append(line.split(",")[8])
    bearing.append(line.split(",")[9])
    #add rsrtip() to get rid of "\n" at the end of range_sm
    range_sm.append(line.split(",")[10].rstrip())
```

```python
#subset to Mandatory & Significant levels

#creating empty lists to hold values from Manditory (4) and Significant (5) levels
station_subset=[]
validUTC_subset=[]
levelcode_subset=[]
pressure_mb_subset=[]
height_m_subset=[]
tmpc_subset=[]
dwpc_subset=[]
drct_subset=[]
speed_kts_subset=[]
bearing_subset=[]
range_sm_subset=[]

#looping by indicies through each level in the sounding to check if it is a Mandatory (4) or Significant (5) level
for i in range(len(levelcode)):
    #if level code is Mandatory (4) or Significant (5)
    if levelcode[i]=="4" or levelcode[i]=="5":
        #add the variable at that level to each list
        station_subset.append(station[i])
        levelcode_subset.append(levelcode[i])
        pressure_mb_subset.append(pressure_mb[i])
        height_m_subset.append(height_m[i])
        tmpc_subset.append(tmpc[i])
        dwpc_subset.append(dwpc[i])
        drct_subset.append(drct[i])
        speed_kts_subset.append(speed_kts[i])
        bearing_subset.append(bearing[i])
        range_sm_subset.append(range_sm[i])
```

```python
#converting strings

#loop through each level in subset of 4s and 5s
for i in range(len(levelcode_subset)):
    #convert data from i-th level from a string to an int/float number
    #Station and UTC are fine as strings
    levelcode_subset[i]=int(levelcode_subset[i])
    if pressure_mb_subset[i]=='M':
        pressure_mb_subset[i]=float('NaN')
    else:
        pressure_mb_subset[i]=float(pressure_mb_subset[i])
        
    if height_m_subset[i]=='M':
        height_m_subset[i]=float('NaN')
    else:
        height_m_subset[i]=float(height_m_subset[i])
        
    if tmpc_subset[i]=='M':
        tmpc_subset[i]=float('NaN')
    else:
        tmpc_subset[i]=float(tmpc_subset[i])
        
    if dwpc_subset[i]=='M':
        dwpc_subset[i]=float('NaN')
    else:
        dwpc_subset[i]=float(dwpc_subset[i])
        
    if drct_subset[i]=='M':
        drct_subset[i]=float('NaN')
    else:
        drct_subset[i]=float(drct_subset[i])
        
    if speed_kts_subset[i]=='M':
        speed_kts_subset[i]=float('NaN')
    else:
        speed_kts_subset[i]=float(speed_kts_subset[i])
        
    if bearing_subset[i]=='M':
        bearing_subset[i]=float('NaN')
    else:
        bearing_subset[i]=float(bearing_subset[i])
        
    if range_sm_subset[i]=='M':
        range_sm_subset[i]=float('NaN')
    else:
        range_sm_subset[i]=float(range_sm_subset[i])
```

```python
#adding units

#assigning specific units to given lists
pressure_mb_subset=pressure_mb_subset*units('millibar')
height_m_subset=height_m_subset*units('meters')
tmpc_subset=tmpc_subset*units('degC')
dwpc_subset=dwpc_subset*units('degC')
drct_subset=drct_subset*units('degrees')
speed_kts_subset=speed_kts_subset*units('knots')
bearing_subset=bearing_subset*units('degrees')
range_sm_subset=range_sm_subset*units('miles')
```

```python
#convert wind speed and wind direction to u, v wind components

u,v=wind_components(speed_kts_subset,drct_subset)

#convert u, v winds to meters/second units
#see 'https://unidata.github.io/MetPy/latest/tutorials/unit_tutorial.html#converting-units' for more information
u=u.to('m/s')
v=v.to('m/s')
```

```python
#save from txt to csv file

with open('KILN_2019052800_2019052810.csv','w') as outfile:
    outfile.write(f"{station[0]} @ {validUTC[0]}\n")
    outfile.write(f"pressure[mb],height[m],temperature[degC],dewpoint[degC],u-wind[m/s],v-wind[m/s]\n")
    for i in range(len(pressure_mb_subset)):
        #0.2f forces the data to output as 2 decimal place floats
        outfile.write(f"{pressure_mb_subset[i].magnitude:0.2f},")
        outfile.write(f"{height_m_subset[i].magnitude:0.2f},")
        outfile.write(f"{tmpc_subset[i].magnitude:0.2f},")
        outfile.write(f"{dwpc_subset[i].magnitude:0.2f},")
        outfile.write(f"{u[i].magnitude:0.2f},")
        outfile.write(f"{v[i].magnitude:0.2f}\n")
```

```python
#open the sounding file as csv

data=pd.read_csv('KILN_2019052800_2019052810.csv',delimiter=',',skiprows=1)
```

```python
#extract variables and add units

#grabing the first line of the data
title=open('KILN_2019052800_2019052810.csv','r').readlines()[0]

#grabing each variable and adding units
pressure=data['pressure[mb]'].values
pressure=pressure*units('millibars')

height=data['height[m]'].values
height=height*units('meters')

temperature=data['temperature[degC]'].values
temperature=temperature*units('degC')

dewpoint=data['dewpoint[degC]'].values
dewpoint=dewpoint*units('degC')

u_wind=data['u-wind[m/s]'].values
u_wind=u_wind*units('m/s')
#convert u-wind from m/s to kts
u_wind=u_wind.to('knots')

v_wind=data['v-wind[m/s]'].values
v_wind=v_wind*units('m/s')
#convert v-wind from m/s to kts
v_wind=v_wind.to('knots')
```

```python
#plot data on skew-t

#create a figure and give it a figsize of 9x9, with a dpi of 100
fig=plt.figure(figsize=(9,9),dpi=100)
#create Skew-T
skew=SkewT(fig,rotation=45)
#add dry adiabats
skew.plot_dry_adiabats()
#add moist adiabats
skew.plot_moist_adiabats()
#add mixing ratio lines
skew.plot_mixing_lines()
#add temperature in red
skew.plot(pressure,temperature,color='red')
#add dewpoint in blue
plt.plot(dewpoint,pressure,color='blue')
#plot wind barbs
skew.plot_barbs(pressure,u_wind,v_wind)
#set x-axis limits that look good (use your eyes)
skew.ax.set_xlim(-35,35)
#set y-axis limits that include the tropopause but not a ton of the stratosphere
skew.ax.set_ylim(1000,150)
#add axis labels with units
plt.xlabel('Temperature (\N{DEGREE SIGN}C)')
#add a y label with units 
plt.ylabel('Hectopascal (hPa)')
#add a title
plt.title(title)
#to make sure everything looks okay
fig.tight_layout()
#save the skew T as skewT.png
plt.savefig('skewT.png')
```

```python
#plot data on hodograph

#remove nans from u and v
u_hodo=u_wind[~np.isnan(u_wind)]
v_hodo=v_wind[~np.isnan(v_wind)]
#create the hodograph (have the range from 0-100kts with 25kt increment)
fig=plt.figure(figsize=(9,9),dpi=100)
hodo=Hodograph(component_range=100)
hodo.add_grid(increment=25)
#make the wind curve black
hodo.plot_colormapped(u_wind,v_wind,height,color='black')
#add x and y labels
plt.xlabel('knots')
plt.ylabel('knots')
#add a title
plt.title(title)
#to make sure everything looks okay
fig.tight_layout()
plt.show
#save figure as hodo.png
plt.savefig('hodo.png')
```
